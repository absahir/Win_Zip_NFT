```python
  # reading two csv files
data1 = pd.read_csv('GeoFRED_All-Transactions_House_Price_Index_by_County_Index_2000=100.csv')
data2 = pd.read_csv('GeoFRED_Equifax_Subprime_Credit_Population_by_County_Percent.csv')
data3 = pd.read_csv('GeoFRED_Equifax_Subprime_Credit_Population_by_County_Percent.csv')
data4 = pd.read_csv('GeoFRED_Housing_Inventory__Active_Listing_Count_by_County_Level.csv')
data5 = pd.read_csv('GeoFRED_Per_Capita_Personal_Income_by_County_Dollars (1).csv')
data6 = pd.read_csv('GeoFRED_Unemployment_Rate_by_County_Percent (1).csv')
```

```python
big_concat = pd.concat([data1.set_index('Region Code'),data2.set_index('Region Code'), data3.set_index("Region Code"), data4.set_index("Region Code"), data5.set_index("Region Code"), data6.set_index("Region Code")], axis=1, join='inner').reset_index()
```
